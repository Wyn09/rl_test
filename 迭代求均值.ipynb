```python
import numpy as np
```

```python
data_size = 100
X = np.random.normal(0, 1, size=(data_size, ))
index = np.random.randint(0, data_size)
print(X.mean())
```

```python
w = X[index]
for i in range(data_size):
    w = w - (w - X[i]) / (i + 1)
print(w)
```

```python
w = X[index]
for i in range(1000000):
    idx = np.random.randint(0, data_size)
    w = w - (w - X[idx]) / (i + 1)
print(w)
```
